```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for Youngs (lengths in nanometers)
d = 1300.0
lamb = 650.0

# Data for plotting
theta = np.arange(-np.pi/2.0, np.pi/2.0, 0.001)
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
bigI = (np.cos(delta/2.0))**2
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
ax.plot(dovlamsintheta, bigI)
ax.set(xlabel='position (d/lambda)*sin(theta)', ylabel='Intensity (I)',
title='Youngs Double Slit Intensity')
ax.grid()
fig.savefig("youngs_ds.png")
plt.show()
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for Youngs (lengths in nanometers)
d = 1000.0
lamb = 650.0

# Data for plotting
theta = np.linspace(-np.pi/2.0, np.pi/2.0, 1000)
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
bigI = (np.cos(delta/2.0))**2
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
ax.plot(dovlamsintheta, bigI, color='red')
ax.set(xlabel='position (d/lambda)*sin(theta)', ylabel='Intensity (I)',
title='Youngs Double Slit Intensity')
ax.grid()

fig.text(.95, .8,"Note: theta=np.linspace(-np.pi/2.0, np.pi/2.0, 1000)")
fig.text(.95, 0.75, "was used to create the graph in place of arange")

#ADD TEXT BELOW
txt = "This is the graph of the intensity pattern that is created by red light (650 nm wavelength)"
txt2 = "and 2 slits that are infinitessimally tiny in width. The slits are placed 1.0μm apart"
txt3 = "and both the zero and first order peaks are shown."
fig.text(.5, -.1, txt, ha='center')
fig.text(.5, -.15, txt2, ha='center')
fig.text(.5, -.2, txt3, ha='center')

fig.savefig("youngs_ds.png")
plt.show()
```

```

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for Youngs (lengths in nanometers)
d = 1000.0
lamb = 650.0

# Data for plotting
theta = np.linspace(-np.pi/2.0, np.pi/2.0, int(np.pi*1000)+1)
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)  #the delta we use for these slit problems   
bigI = (np.cos(delta/2.0))**2    #Intensity just for two slits

fourbigI = ((np.sin(4*delta/2)/np.sin(delta/2))**2)/16
sixteenbigI = ((np.sin(16*delta/2)/np.sin(delta/2))**2)/(16**2)
twofivesixbigI = ((np.sin(256*delta/2)/np.sin(delta/2))**2)/(256**2)

dovlamsintheta = (d / lamb) * np.sin(theta) 
fig, ax = plt.subplots()

ax.plot(dovlamsintheta, sixteenbigI, color='blue') #16
ax.plot(dovlamsintheta, twofivesixbigI, color='black') #256
ax.plot(dovlamsintheta, fourbigI, color='red') #four
ax.set(xlabel='position (d/lambda)*sin(theta)', ylabel='Intensity (I)',
title='Youngs Double Slit Intensity (for N=4, N=16, and N=256)')
ax.grid()
plt.legend(['N=4','N=16','N=256'],bbox_to_anchor = [1,1])
fig.text(.95, .25,"Note: The intensity of N=4, N=16, and N=256")
fig.text(.95, 0.2, "were each scaled by factors of (1/16), (1/16)^2")
fig.text(.95, 0.15, "and (1/16)^4 respectively for better comparison.")

#ADD TEXT BELOW
txt = " These are the intensity pattern graphs produced by red light (650 nm wavelength)"
txt2 = "passing through 4, 16, and 256 slits that are infinitessimally tiny in width."
txt3 = "The slits are placed 1.0μm apart from each other for each respective slit arrangement."
fig.text(.5, -.1, txt, ha='center')
fig.text(.5, -.15, txt2, ha='center')
fig.text(.5, -.2, txt3, ha='center')

fig.savefig("varying_slit_distance.png")
plt.show()
```
